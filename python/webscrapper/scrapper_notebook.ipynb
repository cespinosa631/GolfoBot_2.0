````python
response = requests.get(url+game_modes[0])

html = response.text

print(html[:500])
````

````python
# clean html
soup = BeautifulSoup(html, "html.parser")

# Remove scripts/styles
for tag in soup(["script", "style"]):
    tag.decompose()

clean_text = soup.get_text(separator="\n")

print(clean_text)
````

````python
genai.configure(api_key=GEMINI_API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")
````

````python
prompt = f"""
Extract Age of Empires 3 player statistics from this webpage text.

Return ONLY valid JSON.

Fields:
- player_name
- elo_1v1
- elo_team
- top_1v1_civs
- top_team_civs
- total_games

Webpage text:
{clean_text}
"""

response = model.generate_content(prompt)

print(response.text)
````
